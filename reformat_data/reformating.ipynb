```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import math
```

```python
df = pd.read_excel(r'C:\Users\Eli Brignac\OneDrive\Desktop\Hydrogen\given_files\Parsed_Data.xlsx')
df = df.replace('\n', '', regex=True)
df = df.replace('_x000D_', '', regex=True)
df = df.replace('•', ' ', regex=True)
df.head()
```

```python
df_grouped = df.groupby(['Value Chain', 'Technology', 'Core Occupation'])['Unique Requirements for Hydrogen'].agg(lambda x: '<space> '.join(set(x))).reset_index()
df_grouped
```

```python
df_grouped['Unique Requirements for Hydrogen'] = df_grouped['Unique Requirements for Hydrogen'].str.replace('\n', '')
df_grouped['Unique Requirements for Hydrogen'] = df_grouped['Unique Requirements for Hydrogen'].str.replace(',', '')
df_grouped['Unique Requirements for Hydrogen']
```

```python
df_grouped.to_csv(r'C:\Users\Eli Brignac\OneDrive\Desktop\Hydrogen\df_grouped.csv', index = False)
df_grouped.describe()
```

```python

new_rows = []
def process_row(row):
    if 'Also known as:' in row['Core Occupation']:
        other_names = row['Core Occupation'].split('Also known as:')[1]
        other_names = other_names.split('  ')
        for name in other_names:
            new_row = row.copy()
            new_row['Core Occupation'] = name
            new_rows.append(new_row)
        row['Core Occupation'] = row['Core Occupation'].split('Also known as:')[0]
    return row


for index, row in df_grouped.iterrows():
    process_row(row)

df_grouped = df_grouped.append(new_rows)
df_grouped = df_grouped.drop_duplicates()
df_grouped = df_grouped.reset_index(drop=True)
df_grouped = df_grouped.dropna()
df_grouped = df_grouped[df_grouped['Core Occupation'] != '']
df_grouped.shape
```
