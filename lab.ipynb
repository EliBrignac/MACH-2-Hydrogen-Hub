```python
import pandas as pd
import numpy as np

df = pd.read_csv(r'C:\Users\Eli Brignac\OneDrive\Desktop\Hydrogen\spell_checked.csv')
def append_non_nan(row):
    non_nan_values = [value for value in row[['e', 'f', 'g']] if not pd.isna(value)]
    return row['Unique Requirements for Hydrogen'] + ' '.join(non_nan_values)

df['Unique Requirements for Hydrogen'] = df.apply(append_non_nan, axis=1)

print(df['Unique Requirements for Hydrogen'].iloc[1])
df['Unique Requirements for Hydrogen'].iloc[18]

df = df.drop(['e', 'f', 'g'], axis=1)
df.replace('  ', ' ', regex=True, inplace=True)
df['Unique Requirements for Hydrogen'].iloc[18]
```

```python
df.to_csv(r'C:\Users\Eli Brignac\OneDrive\Desktop\Hydrogen\comma_fixed.csv', index=False)
```
